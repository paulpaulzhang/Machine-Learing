```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
boston = datasets.load_boston()
```

```python
boston.feature_names
```

```python
x = boston.data[:, 5]
y = boston.target
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x, y)
plt.show
```

```python
x = x[y < 50.0]
y = y[y < 50.0]
```

```python
plt.scatter(x, y)
```

```python
from playML.model_selection import train_test_split
from playML.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error
from playML.SimpleLinearRegression import SimpleLinearRegression
```

```python
x_test, y_test, x_train, y_train = train_test_split(x, y, seed=666)
```

```python
reg = SimpleLinearRegression()
```

```python
y_predict = reg.fit(x_train, y_train).predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_predict, color="r")
plt.show()
```

```python
mean_squared_error(y_test, y_predict)
```

```python
root_mean_squared_error(y_test, y_predict)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
from playML.metrics import r2_score
```

```python
r2_score(y_test, y_predict)
```

```python
reg.score(x_test, y_test)
```
