### 测试模型

```python
import pandas as pd
import jieba as jb
import numpy as np
```

```python
test_data = pd.read_csv('datasets/test.csv', lineterminator='\n')
review_id = test_data.values[:, 0]
review = test_data.values[:, 1]
```

```python
from playML import naive_bayes as nb
```

```python
nb_clf = nb.NBClassifier()
nb_clf.load_dataset('datasets/train.csv')
X = nb_clf.data
y = nb_clf.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
vocabList = nb.create_vocab_list(X_train)
trainMatrix = []
for x in X_train:
    trainMatrix.append(nb.setOfWords2Vec(vocabList, x))
```

```python
pPos, p1V, p0V = nb_clf.fit(trainMatrix, y_train)
```

```python
p1V
```

```python
temp = []
for i in range(len(y)):
    temp.append(nb_clf.predict(trainMatrix[i], p0V, p1V, pPos))
```

```python
testMatrix = []
relist = nb.cut_review(review)
for r in relist:
    testMatrix.append(nb.setOfWords2Vec(vocabList, r))
```

```python
result = []
for x in testMatrix:
    result.append(nb_clf.predict(x, p0V, p1V, pPos))
```

```python
result
```

```python
data = {'ID': review_id, 'Pred': result}
df = pd.DataFrame(data, columns=['ID', 'Pred'])
print(df)
```

```python
df.to_csv('datasets/result.csv',index=False)
```
