```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
siny
```

```python
cosy = np.cos(x)
cosy
```

```python
plt.plot(x, siny, color="blue", label="sin(x)", linestyle="-.")
plt.plot(x, cosy, label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("ML")
plt.legend()
plt.show()
```
